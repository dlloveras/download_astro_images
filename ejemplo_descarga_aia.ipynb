```python
from descargar_imagenes_clases import aia_downloader
from descargar_imagenes_clases import initial_final_time
import matplotlib.pyplot as plt
```

```python
#Seleccinoamos el rango temporal en el cual vamos a realizar un search de imagenes disponibles. Seguir el formato indicado
t_ini='2010-12-12 10:00:00'
t_fin='2010-12-12 15:00:00'
ini,fin = initial_final_time(t_ini, t_fin, delta_t=50)
#delta_t es el tiempo en minutos que se le restara a t_ini y se le sumara a t_fin con el fin de extender el rango temporal un total de 2*delta_t.
```

```python
#casteamos la clase aia_downloader a la variable aia_images con los tiempos definidos.
#cadence es la cadencia en minutos. por default es 60
#wave es el pasabandas que nos interesa, ejemplo wave=193. por default busca 171, 193 y 211 simultaneamente.
aia_images = aia_downloader(start_time=ini,end_time=fin,cadence=150)#,wave=193)
```

```python
# la funcion search aplica la busqueda. basicamente sigue a Fido.search de sunpy 
aia_images.search()
#funcion display va a plotear un dot por cada imagen encontrada
aia_images.display()
```

```python
#definir la direccion de descarga. en la clase podrian modificar el directorio por default, asi esto no es necesario modificarlo aca.
aia_images.dir_descarga='/data1/Descargas/temporal/'

#por default origin_download_path = False y se genera una carpeta # del pasabanda  y realiza la descarga dentro
# si se elige origin_download_path = True, la descarga se realiza en el directorio definido por la fecha="%Y%m%d", ejemplo: /data1/Descargas/temporal/171/20101212/.
aia_images.origin_download_path = True

#se corre la funcion descarga. Esta funcion utiliza Fido.fetch desde 'VSO'.
aia_images.download()
#comments: no vuelve a descargar archivos ya existentes en el directorio. 
#Hace 1 nueva iteracion intentando descargar aquellos files que haya fallado. No es infalible ya que el mismo error puede persistir.
#Descarga hasta 5 imagenes en simulateneo por default. se podrian poner mas, incluso podriamos paralelizarlo, pero el ancho de banda seria el cuello de botella.
```
