```python
import os
#main_folder = '/gehme/data/soho/lasco/level_05/c2/'
#main_folder = '/gehme/data/stereo/secchi/L0/a/img/cor2/'
main_folder = '/gehme/data/stereo/secchi/L1/b/img/cor2/'
```

```python
folders=os.listdir(main_folder)
print(folders)
```

```python
tamaño = 3000000 #4Mb
extension='.fts'
for carpetas in folders:
    archivos = os.listdir(main_folder+carpetas)
    for archivo in archivos:
        if os.path.isdir(main_folder+carpetas+"/"+archivo) == False:
            if os.path.getsize(main_folder+carpetas+"/"+archivo) < tamaño:
                if archivo.endswith(extension):
                    os.remove(main_folder+carpetas+"/"+archivo)
```
